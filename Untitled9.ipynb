Linear Programming – Set 03, Problem 3

Constraints
- x + y ≤ 5  
- 8x + y ≥ 8  
- x ≥ 0, y ≥ 0  

Objective:
Maximize P = 3x + 2y

1) Corner Points (Intersections inside the feasible set)
Find all intersection points of:
- x + y = 5  
- 8x + y = 8  
- x = 0  
- y = 0  

Then check which satisfy all inequalities.

2) Plot Feasible Region
Plot lines, shade feasible region, and mark vertices.

3) Objective Function
Compute P = 3x + 2y at each vertex.  
Find the vertex giving the maximum P.

4) Built-in Solver Check
Use `scipy.optimize.linprog` to confirm the result.

5) Slack Variable Form
Convert constraints to equality form with slack variables.


```python
import itertools
import numpy as np
import matplotlib.pyplot as plt
from fractions import Fraction

def h1(t): print(f"\n {t}\n")
def h2(t): print(f"\n {t}\n")
def fx(x): return str(Fraction(x).limit_denominator())

h1("Linear Programming – Set 03, Problem 3")

h2("Constraints")
print("x + y ≤ 5")
print("8x + y ≥ 8")
print("x ≥ 0, y ≥ 0")
print("Objective: Maximize P = 3x + 2y")

lines = [
    (1, 1, 5),
    (8, 1, 8),
    (1, 0, 0),
    (0, 1, 0)
]

def intersect(l1, l2):
    A1, B1, C1 = l1
    A2, B2, C2 = l2
    d = A1*B2 - A2*B1
    if abs(d) < 1e-9:
        return None
    x = (C1*B2 - C2*B1) / d
    y = (A1*C2 - A2*C1) / d
    return np.array([x, y])

def feasible(p):
    x, y = p
    return x >= 0 and y >= 0 and x + y <= 5 and 8*x + y >= 8

h2("1) Corner points")
pts = []
for i, j in itertools.combinations(range(len(lines)), 2):
    p = intersect(lines[i], lines[j])
    if p is not None and feasible(p):
        pts.append(p)
vertices = np.unique(np.round(np.array(pts), 12), axis=0)
for v in vertices:
    print(f"({fx(v[0])}, {fx(v[1])})")

def sort_points(points):
    c = points.mean(axis=0)
    ang = np.arctan2(points[:,1]-c[1], points[:,0]-c[0])
    order = np.argsort(-ang)
    return points[order]

ordered = sort_points(vertices)

h2("2) Feasible region plot")
xs = np.linspace(0, 5.2, 400)
y1 = 5 - xs
y2 = 8 - 8*xs
plt.figure(figsize=(6,6))
X, Y = np.meshgrid(xs, np.linspace(0,5.2,400))
mask = (X+Y<=5)&(8*X+Y>=8)&(X>=0)&(Y>=0)
plt.contourf(X,Y,mask,levels=[-0.5,0.5,1.5],alpha=0.3)
plt.plot(xs,y1,label="x + y = 5")
plt.plot(xs,y2,label="8x + y = 8")
plt.axhline(0)
plt.axvline(0)
for (x,y) in ordered:
    plt.plot(x,y,'o')
    plt.text(x+0.05,y+0.05,f"({x:.2f},{y:.2f})")
plt.xlim(0,5.2)
plt.ylim(0,5.2)
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.title("Feasible Region")
plt.show()

h2("3) Objective values")
c = np.array([3,2])
vals = vertices @ c
for v,val in zip(vertices,vals):
    print(f"({fx(v[0])}, {fx(v[1])}) -> P = {val}")
i = np.argmax(vals)
print(f"Maximum at ({fx(vertices[i][0])}, {fx(vertices[i][1])}) with P = {vals[i]}")

h2("4) Solver check")
try:
    from scipy.optimize import linprog
    A_ub = np.array([[1,1],[-8,-1]])
    b_ub = np.array([5,-8])
    res = linprog(c=-c, A_ub=A_ub, b_ub=b_ub, bounds=[(0,None),(0,None)], method="highs")
    print("Optimal (x,y):", res.x)
    print("Optimal P:", float(c @ res.x))
except:
    print("Install scipy to verify")

h2("5) Slack-variable form")
print("x + y + s1 = 5, s1 ≥ 0")
print("-8x - y + s2 = -8, s2 ≥ 0")
print("x, y ≥ 0")
```

Linear Programming – Set 04, Problem 6 (Snowboards)

Variables
- d = number of Deluxe models  
- s = number of Standard models  

Constraints
- 20d + 10s ≤ 700  →  2d + s ≤ 70  
- 72d + 60s ≤ 3000 →  6d + 5s ≤ 250  
- d ≥ 0, s ≥ 0  

Objective
Maximize revenue R = 324d + 180s  

1) Corner Points
Find all intersection points of:
- 2d + s = 70  
- 6d + 5s = 250  
- d = 0  
- s = 0  
and check which are feasible.

2) Feasible Region
Plot the lines and shade the feasible region.

3) Objective Function
Calculate R = 324d + 180s at each vertex.  
Find the maximum value.

4) Solver Verification
Use `scipy.optimize.linprog` to confirm the result.

5) Slack-Variable Form
Convert constraints into equalities with slack variables:
- 2d + s + u1 = 70, u1 ≥ 0  
- 6d + 5s + u2 = 250, u2 ≥ 0  
- d, s ≥ 0


```python
import itertools
import numpy as np
import matplotlib.pyplot as plt
from fractions import Fraction

def h1(t): print(f"\n {t}\n")
def h2(t): print(f"\n {t}\n")
def fx(x): return str(Fraction(x).limit_denominator())

h1("Set 04 – Problem 6: Snowboard Production")

h2("Constraints and Objective")
print("2d + s ≤ 70")
print("6d + 5s ≤ 250")
print("d ≥ 0, s ≥ 0")
print("Maximize R = 324d + 180s")

lines = [
    (2, 1, 70),
    (6, 5, 250),
    (1, 0, 0),
    (0, 1, 0)
]

def intersect(l1, l2):
    A1, B1, C1 = l1
    A2, B2, C2 = l2
    d = A1*B2 - A2*B1
    if abs(d) < 1e-9:
        return None
    x = (C1*B2 - C2*B1) / d
    y = (A1*C2 - A2*C1) / d
    return np.array([x, y])

def feasible(p):
    d, s = p
    return d >= 0 and s >= 0 and 2*d + s <= 70 and 6*d + 5*s <= 250

h2("1) Corner points")
pts = []
for i, j in itertools.combinations(range(len(lines)), 2):
    p = intersect(lines[i], lines[j])
    if p is not None and feasible(p):
        pts.append(p)
vertices = np.unique(np.round(np.array(pts), 12), axis=0)
for v in vertices:
    print(f"({fx(v[0])}, {fx(v[1])})")

def order_points(points):
    c = points.mean(axis=0)
    ang = np.arctan2(points[:,1]-c[1], points[:,0]-c[0])
    return points[np.argsort(-ang)]

ordered = order_points(vertices)

h2("2) Feasible region plot")
d = np.linspace(0, 42, 400)
s1 = 70 - 2*d
s2 = (250 - 6*d)/5
plt.figure(figsize=(6,6))
D, S = np.meshgrid(np.linspace(0, 42, 400), np.linspace(0, 70, 400))
mask = (2*D + S <= 70) & (6*D + 5*S <= 250) & (D >= 0) & (S >= 0)
plt.contourf(D, S, mask, levels=[-0.5, 0.5, 1.5], alpha=0.3)
plt.plot(d, s1, label="2d + s = 70")
plt.plot(d, s2, label="6d + 5s = 250")
plt.axhline(0)
plt.axvline(0)
for (x, y) in ordered:
    plt.plot(x, y, 'o')
    plt.text(x+0.2, y+0.5, f"({x:.2f},{y:.2f})")
plt.xlabel("d")
plt.ylabel("s")
plt.title("Feasible Region")
plt.legend()
plt.show()

h2("3) Revenue values")
c = np.array([324, 180])
vals = vertices @ c
for v, val in zip(vertices, vals):
    print(f"({fx(v[0])}, {fx(v[1])}) -> R = {val}")
i = np.argmax(vals)
print(f"Maximum at ({fx(vertices[i][0])}, {fx(vertices[i][1])}) with R = {vals[i]}")

h2("4) Solver check")
try:
    from scipy.optimize import linprog
    A_ub = np.array([[2,1],[6,5]])
    b_ub = np.array([70,250])
    res = linprog(c=-c, A_ub=A_ub, b_ub=b_ub, bounds=[(0,None),(0,None)], method="highs")
    print("Optimal (d,s):", res.x)
    print("Optimal R:", float(c @ res.x))
except:
    print("Install scipy to verify")

h2("5) Slack-variable form")
print("2d + s + u1 = 70, u1 ≥ 0")
print("6d + 5s + u2 = 250, u2 ≥ 0")
print("d, s ≥ 0")
```
